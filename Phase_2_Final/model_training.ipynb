```python
import os
import cv2
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, random_split
from torch.cuda.amp import GradScaler, autocast
from pathlib import Path
from tqdm import tqdm
import albumentations as A
from albumentations.pytorch import ToTensorV2
import segmentation_models_pytorch as smp
from torch.optim.lr_scheduler import CosineAnnealingWarmRestarts
from sklearn.metrics import matthews_corrcoef
import matplotlib.pyplot as plt

# ==========================================
# 1. CONFIGURATION
# ==========================================
CONFIG = {
    "base_dir": "/content/drive/MyDrive/GlacierHack_practice/Train",
    "save_path": "/content/drive/MyDrive/GlacierHack_practice/glaciernet_v2.pth",
    "epochs": 50,
    "batch_size": 12,
    "lr": 1e-3,
    "num_workers": 2,
    "val_split": 0.2,
    "seed": 42,
    "device": torch.device("cuda" if torch.cuda.is_available() else "cpu")
}

def set_seed(seed):
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)

set_seed(CONFIG['seed'])

# ==========================================
# 2. DATASET
# ==========================================
class GlacierDataset(Dataset):
    def __init__(self, base_dir, transform=None):
        self.base_dir = Path(base_dir)
        # Loads Band1, Band2, Band3, Band4, Band5
        self.band_dirs = [self.base_dir / f"Band{i}" for i in range(1, 6)]
        self.label_dir = self.base_dir / "labels"
        self.ids = sorted([p.stem for p in self.band_dirs[0].glob("*.tif")])
        self.transform = transform

    def __len__(self): return len(self.ids)

    def __getitem__(self, idx):
        img_id = self.ids[idx]
        # Read 5 bands -> Shape (H, W, 5)
        bands = [cv2.imread(str(d / f"{img_id}.tif"), cv2.IMREAD_UNCHANGED).astype(np.float32) for d in self.band_dirs]
        image = np.stack(bands, axis=-1)

        # Read Label
        label = cv2.imread(str(self.label_dir / f"{img_id}.tif"), cv2.IMREAD_UNCHANGED)
        if label.ndim == 3: label = cv2.cvtColor(label, cv2.COLOR_BGR2GRAY)

        # Map labels: 0->0, 85->1, 170->2, 255->3
        mask = np.zeros_like(label, dtype=np.uint8)
        mask[label == 85] = 1; mask[label == 170] = 2; mask[label == 255] = 3

        # Normalize (Robust Min-Max using percentiles)
        p02, p98 = np.percentile(image, 2), np.percentile(image, 98)
        image = np.clip(image, p02, p98)
        image = (image - image.min()) / (image.max() - image.min() + 1e-6)

        if self.transform:
            aug = self.transform(image=image, mask=mask)
            return aug["image"].float(), aug["mask"].long()
        else:
            return torch.tensor(image.transpose(2,0,1)).float(), torch.tensor(mask).long()

train_transform = A.Compose([
    A.HorizontalFlip(p=0.5), A.VerticalFlip(p=0.5), A.RandomRotate90(p=0.5),
    A.Transpose(p=0.5),
    A.OneOf([
        A.GridDistortion(p=1.0), A.ElasticTransform(p=1.0), A.ShiftScaleRotate(p=1.0),
    ], p=0.5),
    A.CoarseDropout(max_holes=8, max_height=32, max_width=32, p=0.3),
    ToTensorV2(),
])
val_transform = A.Compose([ToTensorV2()])

full_dataset = GlacierDataset(CONFIG['base_dir'], transform=None)
val_size = int(len(full_dataset) * CONFIG['val_split'])
train_size = len(full_dataset) - val_size
train_ds, val_ds = random_split(full_dataset, [train_size, val_size])

# Assign transforms
train_ds.dataset.transform = train_transform
val_ds.dataset.transform = val_transform

train_loader = DataLoader(train_ds, batch_size=CONFIG['batch_size'], shuffle=True, num_workers=CONFIG['num_workers'])
val_loader = DataLoader(val_ds, batch_size=CONFIG['batch_size'], shuffle=False, num_workers=CONFIG['num_workers'])

# ==========================================
# 3. GLACIERNET V2 (Residual + SE Attention)
# ==========================================

class SEBlock(nn.Module):
    """Squeeze-and-Excitation Block"""
    def __init__(self, in_ch, r=16):
        super().__init__()
        self.squeeze = nn.AdaptiveAvgPool2d(1)
        self.excitation = nn.Sequential(
            nn.Linear(in_ch, in_ch // r, bias=False),
            nn.ReLU(inplace=True),
            nn.Linear(in_ch // r, in_ch, bias=False),
            nn.Sigmoid()
        )

    def forward(self, x):
        b, c, _, _ = x.size()
        y = self.squeeze(x).view(b, c)
        y = self.excitation(y).view(b, c, 1, 1)
        return x * y

class ResBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, padding=1, bias=False),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, padding=1, bias=False),
            nn.BatchNorm2d(out_ch),
        )
        self.se = SEBlock(out_ch)
        self.shortcut = nn.Sequential()
        if in_ch != out_ch:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_ch, out_ch, 1, bias=False),
                nn.BatchNorm2d(out_ch)
            )

    def forward(self, x):
        out = self.conv(x)
        out = self.se(out)
        out += self.shortcut(x)
        return F.relu(out)

class GlacierNetV2(nn.Module):
    def __init__(self, n_channels=5, n_classes=4):
        super().__init__()
        self.inc = nn.Sequential(
            nn.Conv2d(n_channels, 32, 3, padding=1, bias=False),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True)
        )
        self.down1 = nn.Sequential(nn.MaxPool2d(2), ResBlock(32, 64))
        self.down2 = nn.Sequential(nn.MaxPool2d(2), ResBlock(64, 128))
        self.down3 = nn.Sequential(nn.MaxPool2d(2), ResBlock(128, 256))
        self.down4 = nn.Sequential(nn.MaxPool2d(2), ResBlock(256, 512))

        self.up1 = nn.ConvTranspose2d(512, 256, kernel_size=2, stride=2)
        self.conv1 = ResBlock(512, 256)
        self.up2 = nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2)
        self.conv2 = ResBlock(256, 128)
        self.up3 = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2)
        self.conv3 = ResBlock(128, 64)
        self.up4 = nn.ConvTranspose2d(64, 32, kernel_size=2, stride=2)
        self.conv4 = ResBlock(64, 32)
        self.outc = nn.Conv2d(32, n_classes, kernel_size=1)

    def forward(self, x):
        x1 = self.inc(x)
        x2 = self.down1(x1)
        x3 = self.down2(x2)
        x4 = self.down3(x3)
        x5 = self.down4(x4)

        x = self.up1(x5)
        if x.shape != x4.shape: x = F.interpolate(x, size=x4.shape[2:], mode='bilinear')
        x = torch.cat([x4, x], dim=1)
        x = self.conv1(x)

        x = self.up2(x)
        if x.shape != x3.shape: x = F.interpolate(x, size=x3.shape[2:], mode='bilinear')
        x = torch.cat([x3, x], dim=1)
        x = self.conv2(x)

        x = self.up3(x)
        if x.shape != x2.shape: x = F.interpolate(x, size=x2.shape[2:], mode='bilinear')
        x = torch.cat([x2, x], dim=1)
        x = self.conv3(x)

        x = self.up4(x)
        if x.shape != x1.shape: x = F.interpolate(x, size=x1.shape[2:], mode='bilinear')
        x = torch.cat([x1, x], dim=1)
        x = self.conv4(x)
        return self.outc(x)

print("‚è≥ Initializing GlacierNet V2...")
model = GlacierNetV2(n_channels=5, n_classes=4).to(CONFIG['device'])

# ==========================================
# 4. LOSS & TRAINING
# ==========================================
class KaggleLoss(nn.Module):
    def __init__(self):
        super().__init__()
        self.lovasz = smp.losses.LovaszLoss(mode='multiclass', per_image=True)
        self.tversky = smp.losses.TverskyLoss(mode='multiclass', alpha=0.4, beta=0.6, from_logits=True)
    def forward(self, inputs, targets):
        return 0.4 * self.tversky(inputs, targets) + 0.6 * self.lovasz(inputs, targets)

criterion = KaggleLoss()
optimizer = optim.AdamW(model.parameters(), lr=CONFIG['lr'], weight_decay=1e-2)
scheduler = CosineAnnealingWarmRestarts(optimizer, T_0=10, T_mult=2, eta_min=1e-6)
scaler = GradScaler()

# HISTORY TRACKING
history = {'train_loss': [], 'val_mcc': []}
best_mcc = -1.0

print("üöÄ Starting Training...")

for epoch in range(CONFIG['epochs']):
    model.train()
    running_loss = 0

    # Training Loop
    for imgs, masks in tqdm(train_loader, desc=f"Ep {epoch+1}"):
        imgs, masks = imgs.to(CONFIG['device']), masks.to(CONFIG['device']).long()

        optimizer.zero_grad()
        with autocast():
            out = model(imgs)
            loss = criterion(out, masks)

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()
        running_loss += loss.item()

    scheduler.step()

    # Store Train Loss
    epoch_loss = running_loss / len(train_loader)
    history['train_loss'].append(epoch_loss)

    # Validation Loop
    model.eval()
    preds, targets = [], []
    with torch.no_grad():
        for imgs, masks in val_loader:
            imgs = imgs.to(CONFIG['device'])
            with autocast():
                # TTA (Test Time Augmentation)
                out = (model(imgs) + torch.flip(model(torch.flip(imgs, [3])), [3])) / 2.0
            preds.append(out.argmax(1).cpu())
            targets.append(masks.cpu())

    mcc = matthews_corrcoef(torch.cat(targets).numpy().flatten(), torch.cat(preds).numpy().flatten())
    history['val_mcc'].append(mcc)

    print(f"Ep {epoch+1} | Loss: {epoch_loss:.4f} | MCC: {mcc:.4f}")

    if mcc > best_mcc:
        best_mcc = mcc
        torch.save(model.state_dict(), CONFIG['save_path'])
        print(f"‚úÖ Saved Best Model! (MCC: {best_mcc:.4f})")

print(f"\nüèÅ Done. Best MCC: {best_mcc:.4f}")

# ==========================================
# 5. VISUALIZATION (PLOTS & SAMPLES)
# ==========================================
print("\nüìä Generating Plots...")

#  Plot Loss and MCC Curves
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
plt.plot(range(1, CONFIG['epochs'] + 1), history['train_loss'], 'r-', label='Train Loss', marker='o', markersize=3)
plt.title('Training Loss vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid(True, alpha=0.3)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(range(1, CONFIG['epochs'] + 1), history['val_mcc'], 'b-', label='Val MCC', marker='o', markersize=3)
plt.title('Validation MCC vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('MCC')
plt.grid(True, alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()

# Visualize Sample Predictions
def visualize_predictions(model, loader, num_samples=3):
    model.eval()
    count = 0

    # Create subplots: Rows = num_samples, Cols = 3 (Image, Truth, Pred)
    fig, axes = plt.subplots(num_samples, 3, figsize=(15, 5*num_samples))
    if num_samples == 1: axes = np.expand_dims(axes, 0)

    with torch.no_grad():
        for imgs, masks in loader:
            imgs = imgs.to(CONFIG['device'])

            # Predict
            with autocast():
                out = (model(imgs) + torch.flip(model(torch.flip(imgs, [3])), [3])) / 2.0
                preds = out.argmax(1).cpu().numpy()

            imgs_np = imgs.cpu().numpy()
            masks_np = masks.numpy()

            for i in range(len(imgs_np)):
                if count >= num_samples:
                    plt.tight_layout()
                    plt.show()
                    return

                # EXTRACT RGB (Bands 4, 3, 2)
                r = imgs_np[i][3, :, :]
                g = imgs_np[i][2, :, :]
                b = imgs_np[i][1, :, :]

                img_vis = np.stack([r, g, b], axis=2) # (H, W, 3)

                # Re-normalize for display brightness
                img_vis = (img_vis - img_vis.min()) / (img_vis.max() - img_vis.min())

                # Plot Input
                axes[count, 0].imshow(img_vis)
                axes[count, 0].set_title(f"RGB (Bands 4,3,2) - Sample {count+1}")
                axes[count, 0].axis('off')

                # Plot Ground Truth
                axes[count, 1].imshow(masks_np[i], cmap='jet', vmin=0, vmax=3)
                axes[count, 1].set_title("Ground Truth")
                axes[count, 1].axis('off')

                # Plot Prediction
                axes[count, 2].imshow(preds[i], cmap='jet', vmin=0, vmax=3)
                axes[count, 2].set_title("Predicted Mask")
                axes[count, 2].axis('off')

                count += 1

print("\nüé® Visualizing 3 Validation Samples (RGB Composition)...")
model.load_state_dict(torch.load(CONFIG['save_path'], map_location=CONFIG['device']))
visualize_predictions(model, val_loader, num_samples=3)
```
